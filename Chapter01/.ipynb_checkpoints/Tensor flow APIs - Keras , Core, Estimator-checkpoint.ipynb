# Tensor flow APIs - Keras , Core, Estimator
TensorFlow is an open source software library developed by Google for numerical computation using data flow graphs. The R interface to TensorFlow is developed by Rstudio which provide an interface to 3 APIs of TensorFLow

- Keras
- Estiamtor
- Core

R packages "keras", "tfestimators" and "tensorflow" provide an R interface to the three APIs respectively. Keras and estimator are high-level APIs, and Core API is a low level which offers full access to core TensorFlow.
TensorFlow supports running computations on a variety of types of devices, including CPU and GPU. To run the GPU version of TensorFlow, we need  NVIDIA® GPU, and then we need to install software components CUDA Toolkit v9.0, NVIDIA® drivers, and cuDNN v7.0 in the system.

## Tensorflow Keras API  ##

Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow, CNTK, or Theano.The R interface to Keras uses TensorFlow™ as it’s default tensor backend engine ."keras" package provides an R interface to TensorFlow Keras API, it lets you build deeplearning models in two manners, sequential and functional. These two ways are discussed in the following sections.

### Sequential API ###
Keras's sequential API is straightforward to understand and implement.It lets you create a neural network linearly, i.e. you build a neural network layer-by-layer where we first initialise a sequential model, and then we stack a series of hidden and output layers on it.

### Getting ready ###

```r
library(keras)
```

```r
# Simulating dummy data
x_data <- matrix(rnorm(1000*784), nrow = 1000, ncol = 784)
y_data <- matrix(rnorm(1000), nrow = 1000, ncol = 1)
```

```r
dim(x_data)
```

```r
dim(y_data)
```

### How to do it..

```r
# Step 1
# Define sequential model
model_sequential <- keras_model_sequential()

# Step 2
# Adding layers to the model
model_sequential %>% 
  layer_dense(units = 16,batch_size =  ,input_shape = c(784)) %>% 
  layer_activation('relu') %>% 
  layer_dense(units = 1)

# Step 3
# complie model
model_sequential %>% compile(
    loss = "mse",
    optimizer = optimizer_sgd(),
    metrics = list("mean_absolute_error")
)
```

```r
# Step 4 summary of model
model_sequential %>% summary()
```

```r
# Step 5: Train the model and store training stats
history <- model_sequential %>% fit(
    x_data, 
    y_data,
    epochs = 30,
    batch_size = 128, 
    validation_split = 0.2
)
```

```r
plot(history)
```

### How it works

We initialise a sequential model in R by calling keras_model_sequential() function, and then stack hidden and output layers on it by using a series of layer functions. In step 2 we add layers to our sequential model, layer_dense() adds a densely-connected NN layer to the defined model. The first layer of the sequential model needs to know what input shape it should expect, and we do this by passing a value to an input_shape argument of the first layer, in our case input shape is equal to the number of features in your data set.

When we add layers to the sequential model keras model object is modified in-place, and we don’t need to assign the updated object back to the original name. This behaviour of keras object is unlike most R objects (R objects are typically immutable).

In our model we have used "relu" activation function, layer_actiavtion() function creates an activation layer which takes input from previous hidden layer and applies activation to the output of our previous hidden layer.We can also use different functions like "leaky_relu", "softmax" etc (activation functions are discussed in other section). The output layer of our model does not use any activation function because we are solving a regression problem.

We can also implement various activation function to each layer by passing a value to argument "activation" in layer_dense() function instead of adding activation layer explicitly. It applies the operation:<br>
**output = activation(dot(input, kernel) + bias)**<br>
where activation is the element-wise activation function passed as the activation argument, kernel is a weights matrix created by the layer, and bias is a bias vector produced by the layer

To train a model, we need to configure the learning process, and this is done in step 3 via the compile() function. In our training process, we apply stochastic gradient descent optimiser to find weights and biases that minimise our objective loss function, i.e. mean squared error." metrics" argument calculates the metric to be evaluated by the model during training and testing.

model_sequential %>% summary() gives summary of the model , it give information about each layer like shape of the output of each layer and parameters in each layer.

In the last step, we train our model for a fixed number of iterations on the dataset, "epochs" argument defines the number of iteration."validation_split" can take float values between 0 and 1. It specifies the fraction of the training data to be used as validation data and "batch_size" defines the number of samples that will be propagated through the network.



### Then there is more
Model progress can be saved during—and after—training. A saved model contains the weight values, the model’s configuration, and the optimiser’s configuration if somehow training process was interrupted we can pick-up training from there.

Below it is shown how we can save the model that we created earlier in section how to do it.

```r
## Save model
model_sequential %>% save_model_hdf5("my_model.h5")
```

If we want to save model after each iteration while training we need to create a checkpoint object. callback_model_checkpoint() is a callback function that performs this task, value of argument "filepath" defines the name of the file that we save in each iteration. It can contain the value of epoch and keys in logs. For example: if filepath is "weights.{epoch:02d}-{val_loss:.2f}.hdf5", then the model checkpoints will be saved with the epoch number and the validation loss in the filename. We have saved our intermediate models file with name as the epoch number in the HDF5 format

The checkpoint object is then passed as value to "callbacks" argument of fit function.

```r
checkpoint_dir <- "checkpoints"
dir.create(checkpoint_dir, showWarnings = FALSE)
filepath <- file.path(checkpoint_dir, "{epoch:02d}.hdf5")

# Create checkpoint callback
cp_callback <- callback_model_checkpoint(
  filepath = filepath,
  verbose = 1
)

# fit model and save model after each check point 
model_sequential %>% fit(
    x_data, 
    y_data,
    epochs = 30,
    batch_size = 128, 
    validation_split = 0.2,
    callbacks = list(cp_callback)
)
```

### Functional API ###
Keras functional API  gives more flexibility to make more complex models, which includes non-sequential connections, multiple inputs/outputs or models with shared layers or models that re-use layers.
In below section we will show an example of multi output model.

### How to do it

```r
# Step 1: Input layer
inputs <- layer_input(shape = c(784))
```

```r
# Step 2: We create  two outputs
predictions1 <- inputs %>%
    layer_dense(units = 8)%>%
    layer_activation('relu') %>%
    layer_dense(units = 1,name = "pred_1")


predictions2 <- inputs %>%
    layer_dense(units = 16)%>%
    layer_activation('tanh') %>%
    layer_dense(units = 1,name = "pred_2")
```

```r
# Step 3
model_functional = keras_model(inputs = inputs,outputs = c(predictions1,predictions2))
```

```r
# Summary of model
summary(model_functional)
```

```r
# Step 4 compile
model_functional %>% compile(
    loss = "mse",
    optimizer = optimizer_rmsprop(),
    metrics = list("mean_absolute_error")
)
```

```r
# Step 5 Train
history_functional <- model_functional %>% fit(
    x_data, 
    list(y_data,y_data),
    epochs = 30,
    batch_size = 128, 
    validation_split = 0.2
)
```

```r
?plot
```

```r
# Plot the model loss of the prediction 1 training data

plot(history_functional$metrics$pred_1_loss, main="Model Loss", xlab = "epoch", ylab="loss", col="blue", type="l")

# Plot the model loss of the prediction 1 validation data
lines(history_functional$metrics$val_pred_1_loss, col="green")

# Plot the model loss of the prediction 2 training data
lines(history_functional$metrics$pred_2_loss, col="red")

# Plot the model loss of the prediction 2 validation data
lines(history_functional$metrics$val_pred_2_loss, col="black")

# Add legend
legend("topright", c("training loss prediction 1","validation loss prediction 1","training loss prediction 2","validation loss prediction 2"), col=c("blue", "green","red","black"), lty=c(1,1))
```

```r
# Plot the model mean squared error of the prediction 1 training data
plot(history_functional$metrics$pred_1_mean_absolute_error, main="Mean Absolute Error", xlab = "epoch", ylab="loss", col="blue", type="l")

# Plot the model mean squared error of the prediction 1 validation data
lines(history_functional$metrics$val_pred_1_mean_absolute_error, col="green")

# Plot the model mean squared error of the prediction 2 training data
lines(history_functional$metrics$pred_2_mean_absolute_error, col="red")

# Plot the model mean squared error of the prediction 2 validation data
lines(history_functional$metrics$val_pred_2_mean_absolute_error, col="black")

# Add legend
legend("topright", c("training mean absolute error prediction 1","validation mean absolute error prediction 1","training mean absolute error prediction 2","validation mean absolute error prediction 2"), col=c("blue", "green","red","black"), lty=c(1,1))
```

### How it works

To create a model using functional API we need to build input and output layer independently, and then pass them to a model function to define the complete model. In the above section, we have created a model with two different output layers which share an input layer/tensor.

In step 1 we have created an input tensor using layer_input() function, which is an entry point into a computation graph generated by Keras model. In next step defined two different output layers. These output layers have different configurations, i.e. activation functions and the number of perceptron units. The input tensor flows through these and produces two different outputs.

In step 3 we define our model using keras_model() function it takes two arguments "inputs" and "outputs" which specifies which layers act as input and output to the model, in case of multi-input or multi-output models, you can use lists of input layers and output layers as shown below:

keras_model(inputs= c(input_layer_1, input_layer_2), outputs= c(output_layer_1, output_layer_2))

Once we have defined our model, in the next steps, we have defined learning process, trained our model and visualised loss and accuracy metrics.compile() and fit() functions used in steps 4 and 5 have been described in details in section how it works of sequential api

### Then there is more


There are scenarios when you want the output of one model to feed into another model along with another input. layer_concatenate() takes as input a list of tensors, all of the same shape, and returns a single tensor, the concatenation of all inputs.

Let's define a new input that we will concatenate with output layer "predictions1" defined earlier.

```r
# Define  new input of the model 
new_input <- layer_input(shape = c(5), name = "new_input")

# Define output layer of new model
main_output <- layer_concatenate(c(predictions1, new_input)) %>%  
  layer_dense(units = 64, activation = 'relu') %>%
  layer_dense(units = 1, activation = 'sigmoid', name = 'main_output')

# We deifine a multi input and multi output model
model <- keras_model(
  inputs = c(inputs, new_input), 
  outputs = c(predictions1, main_output)
)
```

```r
summary(model)
```

## Tensorflow Estimator API ##

Estimators is a high-level TensorFlow API which makes developing deep learning model much easier, with estimators you can write models with high-level intuitive code. It builds the computation graph and provides an environment where we can initialise variables, load data, handle exceptions and create checkpoints.

The "tfestimators" package is an R interface to TensorFlow Estimator API. It implements different model types like linear models and deep neural networks(DNN classifier and regressor). It does not have a direct implementation of recurrent neural networks, convolutional neural networks, support vector machines, random forest, KMeans clustering, etc. but it supports flexible framework for defining arbitrary new model types called as custom estimators.

### Getting ready

```r
# install.packages("tfestimators")
```

```r
library(tfestimators)
```

```r
# training data
x_data_df <- as.data.frame( matrix(rnorm(1000*784), nrow = 1000, ncol = 784))
y_data_df <- as.data.frame(matrix(rnorm(1000), nrow = 1000, ncol = 1))
```

```r
colnames(y_data_df)<- c("target")
dummy_data_estimator <- cbind(x_data_df,y_data_df)
```

### How to do it

```r
# Step 1
features_set <- setdiff(names(dummy_data_estimator), "target")
```

```r
# construct feature columns
feature_cols <- feature_columns(
     column_numeric(features_set)
)
```

```r
# Step 2
# construct input function 
estimator_input_fn <- function(data_,num_epochs = 1) {
  input_fn(data_, features = features_set, response = "target",num_epochs = num_epochs )
}
```

```r
# Step 3
# construct classifier
regressor <- dnn_regressor(
    feature_columns = feature_cols,
    hidden_units = c(5, 10, 8),
    label_dimension = 1L,
    activation_fn = "relu"
)
```

```r
# Step 4
# train classifier with training dataset
train(regressor, input_fn = estimator_input_fn(data_ = dummy_data_estimator))
```

```r
# Step 5
# test data
x_data_test_df <- as.data.frame( matrix(rnorm(100*784), nrow = 100, ncol = 784))
y_data_test_df <- as.data.frame(matrix(rnorm(100), nrow = 100, ncol = 1))
```

```r
colnames(y_data_test_df)<- c("target")
dummy_data_test_df <- cbind(x_data_test_df,y_data_test_df)
```

```r
# predict with test dataset
predictions <- predict(regressor, input_fn = estimator_input_fn(dummy_data_test_df), predict_keys = c("predictions"))
```

```r
head(predictions)
```

```r
# evaluate with test dataset
evaluation <- evaluate(regressor, input_fn = estimator_input_fn(dummy_data_test_df))
```

```r
evaluation
```

### How it works
In section how to do it, we have implemented a DNN regressor using a pre-made estimator, above program can be divided into subparts as below  
1. **Define the feature columns**
In step 1 we create a vector of strings, which contains names of our numeric feature columns, and next, we call feature_columns() function which is a constructor function that defines the expected 'shape' of an input tensor and how features should be transformed(numeric or categorical etc.) during modelling.In our case, the shape of input tensor is 784, and all values of input tensor are numerical. We transform numeric features by passing names of numeric columns to column_numeric() function inside feature_columns(). If you have categorical columns in your data which has values like "categogy_x", "categogy_y" and you want to assign integer values 0, 1 to these we can do this by passing names of the categorical columns to column_categorical_with_identity()

2. Write dataset importing functions.
In step two we define how a pre-made estimator will receive data. It is defined by input_fn() function.It turns raw data sources into Tensors and selects feature and response columns. It also and configure how data is drawn during training, i.e. shuffling, batch size, epochs.

3. Instantiate the relevant pre-made Estimator.
In step 3 we instantiate a pre-made deep neural network (DNN) estimator by calling dnn_regressor(). "hidden_units" argument value of the function defines our network, i.e. hidden layers in the network and number of perceptron in each layer. It takes vectors of integers as the argument value. In our model, we have three layers with 5,10 and 8 perceptrons respectively. We have used relu as our activation function. "label_dimension" argument of the dnn_regrssor function defines the shape of regression target per example.

4. Call a training, evaluation method.
In step 4 we train our model, and in the following step, we predict on test dataset and evaluate test performance.

### Then there is more

Estimators provide a utility called run hooks to track training, report progress, request early stopping and more. One such utility is hook_history_saver() function, which lets us save training history every training step. While training an estimator, we pass our run hooks definition to train() functions "hooks" argument as shown below. It saves model progress after every two training iterations and returns saved training metrics.

```r
training_history <- train(regressor,
                          input_fn = estimator_input_fn(data_ = dummy_data_estimator),
                          steps = 16,
                          hooks = list(hook_history_saver(every_n_step = 2))
                        )
```

### How to do it

```r
x_data = matrix(runif(1000*2),nrow = 1000,ncol = 1)
y_data = matrix(runif(1000),nrow = 1000,ncol = 1)

# Creating TF variables and define model
W <- tf$Variable(tf$random_uniform(shape(1L), -1.0, 1.0))
b <- tf$Variable(tf$zeros(shape(1L)))
y_hat <- W * x_data + b

# Define loss finction and Minimize mean squared errors.
loss <- tf$reduce_mean((y_hat - y_data) ^ 2)
optimizer <- tf$train$GradientDescentOptimizer(0.5)
train <- optimizer$minimize(loss)

# Launch graph and initialize the variables.
sess = tf$Session()
sess$run(tf$global_variables_initializer())

# Train the model
for (step in 1:201) {
  sess$run(train)
  if (step %% 20 == 0)
    cat(step, "-", sess$run(W), sess$run(b), "\n")
}

sess$close()
```

### How it works..

TensorFlow programs generate a computation graph in which nodes of the graph are called ops. These ops take tensors as input and perform computations and produce tensors(Tensors are an n-dimensional array or list) TensorFlow programs are structured in two phases, construction phase and execution phase. In the construction phase, we assemble the graph, and in the execution phase, we execute graphs in the context of the session.

Calling tensorflow package in R creates an entry point to TensorFlow API `tf` through which we can access the main Tensorflow module. tf$Variable() function creates a variable which holds and updates trainable parameters.TensorFlow variables are in-memory buffers containing tensors. In the above code, lines 3&4 create two TF variables of Weights and Bias with initial values.
In the next line, we define our model.

In line 6 we define loss function as per below equation:
 $$ {MSE} = {\frac {1}{n}}\sum _{i=1}^{n}(Y_{i}-{\hat {Y_{i}}})^{2}. $$

Function reduce_mean() computes mean of elements across dimensions of a tensor, in our code it calculates average loss over the training set. In tensorflow we can use different optimisation algorithms to train our network. In line 7 we have used gradient descent optimiser with learning rate 0.5. In the next step, we define the objective of our training step i.e. minimise loss.

In lines 3 to 8 we assemble computation graph and now TensorFlow knows the description of computations that we want to execute. In our case we want TensorFlow to minimise loss, i.e. mean squared error using gradient descent algorithm.TensorFlow does not run any computation until the session is created and run() function is called.

In line 9 we launch the session and next we add an ops to run `tf` variable initialisers. sess$run(tf$global_variables_initializer()) initializes all variables in current session parallely. We should run this ops after we have fully constructed the graph and launched it in a session.

Then we run training step 201 times using a for loop and print out tf variables i.e weight and bias every  iteration.
