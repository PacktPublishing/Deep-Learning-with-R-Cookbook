## How to do it

### Step 1

Let us check the datatype of "Sales" column in our data.

```r
class(data$Sales)
```

Note that in the data, "Sales" column is of factor datatype. We need to change this to numeric in order to use it in our analysis.

```r
data$Sales <- as.numeric(as.character(data$Sales))
class(data$Sales)
```

### Step 2

To implement a time series forecasting,we need to convert the data into stationery data.
We can do so by using the diff() function to calculate the iterated difference. We pass the value of argument differences as 1 since we want the differencing with lag =1. 

```r
data_differenced = diff(data$Sales, differences = 1)
head(data_differenced)
```

### Step 3

Next we create a supervised dataset in order to apply GRU.We transform the series "data_differenced" by creating a lag in the series with order of 1,i.e. the value at time (t-1) as the input will have value at time t as the output.

```r
data_lagged = c(rep(NA, 1), data_differenced[1:(length(data_differenced)-1)])
data_preprocessed = as.data.frame(cbind(data_lagged,data_differenced))
colnames(data_preprocessed) <- c( paste0('x-', 1), 'x')
data_preprocessed[is.na(data_preprocessed)] <- 0
head(data_preprocessed)
```

### Step 4

We now split the data into training and testing sets

In time series problems we cannot do random sampling on the data since in these problems order of the data matters.
Thus we split the data by taking first 70% of the  series as training and the remaining last 30% of the data as testing data.

```r
N = nrow(data_preprocessed)
n = round(N *0.7, digits = 0)
train = data_preprocessed[1:n, ]
test  = data_preprocessed[(n+1):N,]
print("Training data snapshot :")
head(train)

print("Testing data snapshot :")
head(test)
```

### Step 5

In this step we normalize the data within the scale of the activation function that we are going to use.Since we are going with tanh as the choice of our activation function which ranges from -1 to +1, we scale the data using min-max normalization.
Here the normalization happens with respect to the training data.

```r
scaling_data = function(train, test, feature_range = c(0, 1)) {
  x = train
  fr_min = feature_range[1]
  fr_max = feature_range[2]
  std_train = ((x - min(x) ) / (max(x) - min(x)  ))
  std_test  = ((test - min(x) ) / (max(x) - min(x)  ))
  
  scaled_train = std_train *(fr_max -fr_min) + fr_min
  scaled_test = std_test *(fr_max -fr_min) + fr_min
  
  return( list(scaled_train = as.vector(scaled_train), scaled_test = as.vector(scaled_test) ,scaler= c(min =min(x), max = max(x))) )
  
}


Scaled = scaling_data(train, test, c(-1, 1))

y_train = Scaled$scaled_train[, 2]
x_train = Scaled$scaled_train[, 1]

y_test = Scaled$scaled_test[, 2]
x_test = Scaled$scaled_test[, 1]
```

We write a function to revert the predicted values to the original scale.We will use this function while predicting the values.

```r
## inverse-transform
invert_scaling = function(scaled, scaler, feature_range = c(0, 1)){
  min = scaler[1]
  max = scaler[2]
  t = length(scaled)
  mins = feature_range[1]
  maxs = feature_range[2]
  inverted_dfs = numeric(t)
  
  for( i in 1:t){
    X = (scaled[i]- mins)/(maxs - mins)
    rawValues = X *(max - min) + min
    inverted_dfs[i] <- rawValues
  }
  return(inverted_dfs)
}
```

### Step 6

In this step we define the model and configure the layers.We reshape our data into a 3-D format so that it can be fed into the model.

```r
# Reshaping the input to 3-dimensional
dim(x_train) <- c(length(x_train), 1, 1)

# specify required arguments
batch_size = 1               
units = 1                     
#=========================================================================================

model <- keras_model_sequential() 
model%>%
  layer_gru(units, batch_input_shape = c(batch_size, dim(x_train)[2], dim(x_train)[3]), stateful= TRUE)%>%
  layer_dense(units = 1)
```

Let us have a look at the summary of the model.

```r
summary(model)
```

Next, we compile the model

```r
model %>% compile(
  loss = 'mean_squared_error',
  optimizer = optimizer_adam( lr= 0.01, decay = 1e-6 ),  
  metrics = c('accuracy')
)
```

### Step 7 

Now in each epoch we fit our training data into the model and reset the states.

```r
for(i in 1:50 ){
  model %>% fit(x_train, y_train, epochs=1, batch_size=batch_size, verbose=1, shuffle=FALSE)
  model %>% reset_states()
}
```

### Step 8

Finally we predict the values for the test dataset and also use the "inverse_scaling" function to scale back the predictions to original scale

```r
scaler = Scaled$scaler
# predictions = numeric(L)
predictions = vector()

for(i in 1:length(x_test)){
     X = x_test[i]
     dim(X) = c(1,1,1)
     yhat = model %>% predict(X, batch_size=batch_size)
     # invert scaling
     yhat = invert_scaling(yhat, scaler,  c(-1, 1))
     # invert differencing
     yhat  = yhat + data$Sales[(n+i)]
     # store
     predictions[i] <- yhat
}
```

Let us see the predictions for the test data

```r
predictions
```

# How it works

Step 1<br>
In the first step, we check the datatype of the column whose values are to be forecasted; in our data, it is the "Sales" column. We change it to numeric datatype.

Step 2<br>
In this step, we convert the input data into stationery data. This is done in order to get rid of time-dependent components in the data. We saw that our input data has an increasing trend. In time series forecasting, it is advised to get rid of the trend component before building the model. These trends can be added back to forecasted values later so that we can get our predictions in the original scale. In this example, we will remove the trend by differencing the data with an order of 1,i.e. by subtracting previous observation from the current one.


Step 3<br>
While using algorithms like LSTMs and GRUs, we need to feed data in supervised form,i.e. in the form of predictors and target variables. In time series problems, we deal with data where the value at time (t-k) will act as the input for value at time t for any k stepped lagged dataset. In our example k = 1 and hence we create a lagged dataset by transforming the current data into a lagged data of order 1. Thus you can see the X=t-1 and Y=t pattern in the data. The lagged dataset series which we create now acts as the predictor variable.


Step 4<br>
We split the data into training and testing datasets. Random sampling does not allow us to maintain the order of the observations in a time series data. Hence we split the data keeping the order of observations intact, here we are taking first 70 % of the data as training and rest 30% as testing data.*n* represents the split point which is basically the last index of training and *n+1* represents the starting index of the testing data.


Step 5<br>

GRUs expect data to be within the scale of the activation function used by the network. Since we are using tanh as the activation function, which gives outputs in the range of (-1,1), we need to scale the training and testing dataset also within the range of (-1,1). We use min-max scaling in this case.
 $$ z_i = \frac{x_i - x_{min}}{x_{max} - x_{min}} $$

Also note that to make the experiment fair, the scaling coefficients (min and max) values must be calculated on the training dataset and applied to scale the test dataset and any forecasts to avoid bias the experiment with any knowledge from the test dataset. That is why here min and max values are used as the reference to scale both training and testing data and also the predicted values.

We also create a function named as *invert_scaling* to do the inverse scaling of scaled values so that we can map the scaled values back to the original scale.


Step 6 <br>
GRUs expect data in a specific format which is of the form : [batch_size, timesteps, features].<br>
- *batch_size* defines the number of observations to be fed to the model in each batch.
*timesteps* represent the number of time steps the model needs to look back in the historical data for prediction. In this example we have set it to 1.
- *features* represent the number of predictors we use , this is 1 in our case.
So in this step we reshape the input data in the required format and feed this into the gru layer.
Note that we have specified the argument *stateful* = TRUE so that the last state for each sample at index i in a 
batch will be used as initial state for the sample of index i in the following batch.This assumes a one-to-one mapping between samples in different successive batches.
     
*units* represent the dimension of the output space.Since we are dealing with predicting continuous values,we take units =1 in this case.


Once we have defined the model we compile it.We specify *mean_squared_error* as the loss function, *adaptive monument estimation(adam)* as the optimization algorithm and learning rate and learning rate decay over each update.
*accuracy* is the metric to assess the model performance here.

Step 7<br>
In the next step, for each epoch, we fit the training data into the model. We specify the value of argument shuffle as false to avoid any kind of shuffling in the training data while building the model. This is done to prevent the time dependencies between the observations. We also reset the network state after each epoch via the function reset_states().
Since we have specified value of the argument *stateful*  = True, we need to manually reset the states of the LSTM after each epoch to have a clean setup for the next one.


Step 8<br>
In the last step, we predict the values for testing dataset. To scale the predicted values back to the original scale, we use the *inverse_scaling* function, which was defined in Step 5.

While working with large datasets, we often run out of memory while training a deep learning model. The keras library in R provides various generator utility functions which generate batches of training data on the fly during the training process. It also provides a utility function to create batches of temporal data. Below code creates a supervised form of data similar to the previous section, Time series forecasting with GRU.

```r
# importing required libraraies
library(reticulate)
library(keras)

# generating dummy data
data =  seq(from = 1,to = 10)

# timseries generator
gen = timeseries_generator(data = data,targets = data,length = 1,batch_size = 5)

# Print first batch of generator
iter_next(as_iterator(gen))
```

Above we see that a generator object yields a list of two sequences, the first sequence is a feature vector, and the second sequence is the corresponding label vector.

Below code implemets a custom generator for time series data. It gives the flexibility to define "lookback" , i.e how many past values we should use to predict the future value or sequence. "future_steps" defines number of future timesteps to predict.

```r
generator <- function (data,lookback =3 ,future_steps = 3,batch_size = 3 ){
    new_data = data
    for(i in seq(1,3)){
        data_lagged = c(rep(NA, i), data[1:(length(data)-i)])
        new_data = cbind(data_lagged,new_data)
    }    
    targets = new_data[future_steps:length(data),(ncol(new_data)-(future_steps-1)):ncol(new_data)]
    gen = timeseries_generator(data = data[1:(length(data)-(future_steps-1))],targets = targets,length = lookback,batch_size = batch_size)
}
```

```r
cat("First batch of generator:")
iter_next(as_iterator(generator(data = data,lookback = 3,future_steps = 2)))
```
